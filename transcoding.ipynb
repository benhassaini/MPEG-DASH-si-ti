```python
def categoriser_resolution(level):
    if level == '320x240' :
        return 1
    elif level == '480x360':
        return 2
    elif  level == '854x480':
        return 3
    elif level =='1280x720':
        return 4
    else:
        return 5
def categoriser_psnr(level):
    if level >= 45:
        return 5
    elif 33 <= level < 45:
        return 4
    elif 27.4 <= level < 33:
        return 3
    elif 18.7 <= level < 27.4:
        return 2
    else:
        return 1
dfresult_all['Resolution_category'] = dfresult_all['resolution'].apply(lambda x: categoriser_resolution(x))
dfresult_all.drop(columns=['resolution'], inplace=True)
dfresult_all['PSNR_Category'] = dfresult_all['PSNR'].apply(lambda x: categoriser_psnr(x))
dfresult_all
```
